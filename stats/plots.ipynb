```python
import matplotlib.pyplot as plt

# Dataset 1: Baseline
deviation1 = [173611, 347222, 279019, 102125, 245099]
ber1 = [5.14, 6.28, 0.45, 1.44, 0.04]

# Dataset 2: Comparison
deviation2 = [173611, 347222, 279019, 102125, 245099]
ber2 = [0.11, 0.25, 2.5, 0.74, 0.007]

# Sort both datasets by deviation
data1_sorted = sorted(zip(deviation1, ber1))
data2_sorted = sorted(zip(deviation2, ber2))

dev1_sorted, ber1_sorted = zip(*data1_sorted)
dev2_sorted, ber2_sorted = zip(*data2_sorted)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(dev1_sorted, ber1_sorted, marker='x', linestyle='-', label='Baseline', color='blue')
plt.plot(dev2_sorted, ber2_sorted, marker='x', linestyle='-', label='ECC', color='orange')

plt.title('BER [%] across deviation [Hz]')
plt.xlabel('Deviation [Hz]')
plt.ylabel('BER [%]')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
