```python
#!/usr/bin/env python3
import matplotlib.pyplot as plt

def main():
    # Data: first column is distance, second column is BER [%]
    distance = [300, 220, 130, 65, 10]
    ber = [3.45804989, 6.15744774, 8.04526356, 0.18552876, 2.73351648]

    # Create a new figure
    plt.figure(figsize=(8, 6))
    
    # Plot the data using markers and a connecting line
    plt.plot(distance, ber, marker='o', linestyle='-', color='blue')
    
    # Add labels and title
    plt.xlabel('Distance')
    plt.ylabel('BER [%]')
    plt.title('Base Case: BER vs Distance')
    
    # Optional: add a grid for easier visualization
    plt.grid(True)
    
    # Show the plot
    plt.show()

if __name__ == '__main__':
    main()
```
